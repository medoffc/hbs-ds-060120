2. List the names of all customers that have a creditLimit between 100,000 and 200,000 inclusively. Order by customer name.

3. List all employees who have the initials of M.P. or P.M., by name. Order by last name.

```python
cur.execute("""SELECT 
    lastname, firstname
FROM
    classicmodels.employees
WHERE
    lastname LIKE 'M%'
        AND firstname LIKE 'P%'
        OR lastname LIKE 'P%'
        AND firstname LIKE 'M%'
ORDER BY lastname""")
df = pd.DataFrame(cur.fetchall())
df. columns = [i[0] for i in cur.description]
df.head()
```

4. List all the products' productNames purchased by Mini Wheels Co. Order the list by product line.

```python
cur.execute("""SELECT 
    productName, productline, customername
FROM
    classicmodels.products
        JOIN
    orderdetails USING (productcode)
        JOIN
    orders USING (ordernumber)
        JOIN
    customers USING (customerNumber)
WHERE
    customername LIKE 'Mini Wheels%'
ORDER BY productline""")
df = pd.DataFrame(cur.fetchall())
df. columns = [i[0] for i in cur.description]
df.head(22)
```

5. List the customers by customerName whose contact person shares either a first name or last name with any of the company's employees Order by customerName

```python
cur.execute("""SELECT DISTINCT
    contactfirstName, contactlastname
FROM
    customers c
        LEFT JOIN
    employees emp_first ON emp_first.firstName = c.contactfirstname
        LEFT JOIN
    employees emp_last ON emp_last.lastname = c.contactlastname
WHERE
    emp_first.firstName IS NOT NULL
        OR emp_last.lastName IS NOT NULL""")
df = pd.DataFrame(cur.fetchall())
df. columns = [i[0] for i in cur.description]
df.head(22)
```

9. Identify the offices (by city) that have the least employees and the most employees (answer should include city, numEmps)

```python
cur.execute("""SELECT 
    city, COUNT(*) AS employeeCount
FROM
    employees
        JOIN
    offices USING (officeCode)
GROUP BY officeCode
HAVING employeeCount IN (SELECT 
        MAX(employeeCount)
    FROM
        (SELECT 
            officeCode, COUNT(*) AS employeeCount
        FROM
            employees
        GROUP BY officeCode) AS max_count UNION SELECT 
        MIN(employeeCount)
    FROM
        (SELECT 
            officeCode, COUNT(*) AS employeeCount
        FROM
            employees
        GROUP BY officeCode) AS min_count)""")
df = pd.DataFrame(cur.fetchall())
df. columns = [i[0] for i in cur.description]
df.head()
```
